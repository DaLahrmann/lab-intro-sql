```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # To get the password without showing the input
```

```python
password = getpass.getpass()
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python

query="SELECT COUNT(DISTINCT last_name) \
    FROM sakila.actor;"
pd.read_sql_query(query, engine)
```

```python
query="SELECT rental_date, \
    MONTH(rental_date) AS 'month', \
    WEEKDAY(rental_date) AS 'weekday', \
    CASE\
        WHEN WEEKDAY(rental_date)>4 THEN 'weekend' \
        ELSE 'workday'\
    END AS 'day_type'\
    FROM sakila.rental;"
pd.read_sql_query(query, engine)
```

```python
query="select *\
    from sakila.film\
    where title like '%ARMAGEDDON%';"
pd.read_sql_query(query, engine)
```

```python
query="select *\
    from sakila.film\
    order by length desc\
    limit 10;"
pd.read_sql_query(query, engine)
```

```python
query="select count(*)\
    from sakila.film\
    where special_features like '%Behind the Scenes%';"
pd.read_sql_query(query, engine)
```

```python
query="select rating, round(avg(length),2) as avg_length\
    from sakila.film\
    group by rating\
    having avg_length > 120\
    order by rating asc;"
pd.read_sql_query(query, engine)
```

```python
query="select title, length,\
        rank() over (\
            order by length desc) as 'rank'\
    from sakila.film\
    where not (length is null) and\
        (length > 0)\
    order by length desc;"
pd.read_sql_query(query, engine)
```
